### Analisis Proyeksi Penduduk di Indonesia Pada Tahun 2023 

##### Data ini merupakan kumpulan indikator strategis tingkat provinsi di Indonesia pada tahun 2023 yang dirilis oleh Badan Pusat Statistik (BPS). Data ini menyajikan gambaran korelasi antara kualitas sumber daya manusia dengan kondisi ekonomi di tiap wilayah. Analisis ini meliputi Import Data dengan Pandas, Split Kolom Data dengan Menggunakan Iloc, Merge, dan Export  Output.

### Import Data

```python
import pandas as pd
data = pd.read_csv(r"C:\Users\Marcella\Downloads\Indeks Pembangunan Manusia Menurut Provinsi, 2023.csv", sep=';')
print(data)
```

### Preview Data 

```python
print("===== DATA ASLI =====")
print(data.head())
```

### Split Kolom dengan Menggunakan ILOC


```python
# iloc_1: Mengambil kolom Provinsi dan IPM (kolom 0 dan 1)
# iloc_2: Mengambil kolom Kemiskinan, TPT, dan PDRB (kolom 2, 3, 4)
df_iloc_1 = data.iloc[:, [0, 1]]
df_iloc_2 = data.iloc[:, [2, 3, 4]]

# Preview Data iloc ke 1 dan iloc ke 2
print("\n--- Preview Data iloc 1 (Provinsi & IPM) ---")
print(df_iloc_1.head())

# Menambahkan Index yang sama dengan iloc 1 ke iloc 2
df_iloc_2.index = df_iloc_1.index

print("\n--- Preview Data iloc 2 (Ekonomi & Sosial) ---")
print(df_iloc_2.head())
```

### Lakukan Merge Data Berdasarkan Index

```python
# Kita gabungkan kembali bagian 1 dan bagian 2
df_merge = pd.merge(df_iloc_1, df_iloc_2, left_index=True, right_index=True)
```

### Preview Data Merge

```python
print("\n--- Preview Data Setelah Merge ---")
print(df_merge.head())
```

### Export Output

```python
path_export = r'C:\Users\Marcella\Downloads\Hasil_Olah_BPS_Marcella.csv'
df_merge.to_csv(path_export, index=False)

print(f"\nSukses! File berhasil disimpan di: {path_export}")
```
