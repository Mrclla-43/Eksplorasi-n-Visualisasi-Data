### CREATE DATAFRAME

```python
import pandas as pd

# Create a sample DataFrame

data = {'Product': ['A', 'B', 'A', 'B', 'C', 'C', 'A'],

'Category': ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Electronics', 'Furniture', 'Furniture'],

'Sales': [250, 150, 350, 200, 100, 400, 300],

'Quantity': [10, 5, 7, 8, 3, 9, 4]}

df = pd.DataFrame(data)

print(df)
```

### CEK TIPE DATA DALAM FRAME

```python
# Mengecek tipe data dari setiap kolom

print(df.dtypes)
```

### CEK TIPE DATA LENGKAP DALAM FRAME

```python
# Mengecek informasi lengkap tentang DataFrame

df.info()
```

### UBAH TIPE DATA DALAM DATAFRAME

```python
# Mengubah tipe data kolom 'Sales' menjadi float

df['Sales'] = df['Sales'].astype(float)

print(df.dtypes)
```

### SUMMARIZE (AGG) -> MENJUMLAHKAN SECARA KESELURUHAN

```python
# Summarizing sales and quantity for each category

summary = df.groupby('Category').agg({'Sales': 'sum', 'Quantity': 'mean'})

print(summary)
```

### GROUPING DATA

```python
# Grouping data by Product and summarizing total sales

grouped = df.groupby('Product').agg({'Sales': 'sum', 'Quantity': 'mean'})

print(grouped)
```

### FILTER (loc)

```python
# Filtering rows where Sales are greater than 200

filtered = df[df['Sales'] > 200]

print(filtered)
```

### MEMILIH KOLOM YG INGIN DIPERLIHATKAN

```python
# Selecting Product and Sales columns

selected_columns = df[['Product', 'Sales']]

print(selected_columns)
```

### MEMBUAT VARIABEL BARU

```python
df['Revenue'] = None

print(df)
```

### MENGURUTKAN DATA (SORTING)

```python
# Sorting by Sales in descending order

sorted_df = df.sort_values('Sales', ascending=False)

print(sorted_df)
```

### PIE CHART

```python
import matplotlib.pyplot as plt

# Pie chart berdasarkan 'Product' dan 'Sales'

plt.figure(figsize=(7, 7))

plt.pie(sorted_df['Sales'], labels=sorted_df['Product'], autopct='%1.1f%%', startangle=90)

plt.title('Sales Distribution by Product')

plt.show()
```

### BAR CHART

```python
# Bar plot untuk 'Product' dan 'Sales'

plt.figure(figsize=(8, 6))

plt.bar(sorted_df['Product'], sorted_df['Sales'], color='skyblue')

plt.title('Sales by Product')

plt.xlabel('Product')

plt.ylabel('Sales')

plt.show()
```

### HISTOGRAM

```python
# Histogram untuk distribusi 'Sales'

plt.figure(figsize=(8, 6))

plt.hist(sorted_df['Sales'], bins=5, color='green', edgecolor='black')

plt.title('Sales Distribution')

plt.xlabel('Sales')

plt.ylabel('Frequency')

plt.show()
```

### SCATTERPLOT

```python
# Scatterplot untuk 'Sales' dan 'Quantity'

plt.figure(figsize=(8, 6))

plt.scatter(sorted_df['Sales'], sorted_df['Quantity'], color='red', s=100)

plt.title('Sales vs Quantity')

plt.xlabel('Sales')

plt.ylabel('Quantity')

plt.show()
```
