### DataFrame

```python
import pandas as pd
# DataFrame dari Dictionary
data_start = {
    "Nama" : ["Marcella", "Marsha", "Mareko", "Michael", "Rafli"],
    "Umur" : [18, 17, 18, 19, 18],
    "Kota" : ["Palembang", "Sumba", "Nias", "Jakarta", "Tangerang"], 
    "Status" : ["Mahasiswa", "Mahasiswa", "Mahasiswa", "Mahasiswa", "Mahasiswa"]
}
df = pd.DataFrame(data_start)
print(df)
```

```python
# DataFrame dari list of lists
data_finish = [
    ["Marcella", 18, "Palembang", "Mahasiswa"],
    ["Marsha", 17, "Sumba", "Mahasiswa"],
    ["Mareko", 18, "Nias", "Mahasiswa"],
    ["Michael", 19, "Jakarta", "Mahasiswa"],
    ["Rafli", 18, "Tangerang", "Mahasiswa"]
]
df2 = pd.DataFrame(data_finish, columns=["Nama", "Umur", "Kota", "Status"])
print(df2)
print("")
print(df.columns.to_list())
print("")
df2.columns=["Name", "Age", "City", "Position"]
print(df2)
```
